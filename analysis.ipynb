### Testing Orientation-Based Analysis of 3D Printing Parameters and Quality Control Dataset

**Source**
Patil, Sonali Sagar; Thite, Sandip; Deshpande, Yogesh; Parle, Dattatraya  (2025), “Orientation-Based Analysis of 3D Printing Parameters and Quality Control”, Mendeley Data, V2, doi: 10.17632/93bdmx2b7p.2

**Description**: 
The provided dataset is a structured tabular dataset used for 3D printing analysis, containing features that describe the characteristics and parameters of 3D printed models. These features include both numerical values and categorical data.

**Dataset Category**: Three Dimensional Printing, Stereolithography

### Dataset Features

---

"This dataset is suitable for various analyses, including statistical analysis, machine learning modeling, and quality control in 3D printing processes. It can help optimize printing parameters, improve print quality, and ensure the structural integrity of printed models." stated in the documentation.

### Numerical Features:

**dX, dY, dZ**:
- represent bounding box dimensions
- define max printible area for 3D model
- *important for determining if a model fits within printer's capacity*

**now** (number of walls):
- total number of vertical walls in the model, regardless of position or height.
- more walls = sturdier build, but require more time and material
- *important for determining overall print model structural complexity and density*

**nowp** (number of walls on print bed):
- subset of "now" walls printed directly on the print bed
- indicates how much of the model has initial contact with the bed
- *important for determining adhesion, stability, and early-layer accuracy of the print*

**Fwall** (First Wall):
- boolean value
- indicates if first wall is printed directly on the print bed
- vital for preventing warping
- *important for understanding initial adhesion*

**Farea** (First Area):
- refers to minimum area of the first layer being printed
- smaller surfaces might struggle to adhere to the print bed
- *important for tracking first layer print failures*

**Offset**:
- difference in the center of gravity between first and last printed layers
- *important for detecting potential shifts or misalignments during printing*

**Maxoffset_bed**:
- indicates maximum difference between the wall's first and last layer when printed on the bed
- *important for tracking stability and uniformity of the print*

**Wanother**:
- number of walls printed close to or on top of each other
- *important for understanding print model structural integrity and density*

**maxoffset_anotherW**:
- indicates maximum difference between walls printed close to or on top of one another
- *important for assessing uniformity and stability of closely placed walls*

**Wair**:
- number of walls printed in the air, without direct support from the layer below
- *important anomaly that can cause issues like sagging or deformation*

**Wair_partial**:
- number of walls partially printed in the air
- *important anomaly that can cause issues like sagging or deformation*

**max_support**:
- indicates largest area of the initial wall layer being printed on the bed
- *important for determining if print model has a strong foundation and good adhesion to print bed*

### Categorial Features:

**model**:
- model ID

**Fwall** (First Wall):
- boolean value
- indicates if first wall is printed diretly on the print bed
- vital for preventing warping
- *important for understanding initial adhesion and print stability*

**needsupport**:
- boolean value
- indicates if a model contains overhangs or geometry that needs supporting material
- indicates a need for post processing the print, which can lead to leftover residual on the model (have to cut supports and sand models)
- *important for preventing unsupported layer deformation or sagging, also affects post-processing effort.*
